# Model Training

```python
#splitting data into input and output variable
x=cd_new.drop("Province_State",axis=1) #separating independent and target variable
y=cd_new["Province_State"]
```

```python
x.shape
```

```python
y.shape
```

```python
#scaling in input variables
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
#Importing all the model library

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score


from sklearn.linear_model import LinearRegression, Lasso, Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#Splittinfg the data into training and testing data
#Splittinfg the data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
#All Algorithm by using for loop
model=[LinearRegression(), Lasso(), Ridge(),ElasticNet(),SVR(),DecisionTreeRegressor()]
for m in model:
    m.fit(x_train,y_train)
    print('Score of',m,'is:',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print('Error:')
    print('Mean Absolute Error :',mean_absolute_error(y_test,predm))
    print('Mean Squared Error :',mean_squared_error(y_test,predm))
    print('Root Mean Squared Error :',np.sqrt(mean_squared_error(y_test,predm)))
    print('r2 score :',r2_score(y_test,predm))
    print('*****************************************************')
    print('\n')
```

```python
#Cross Validate the Models
from sklearn.model_selection import cross_val_score
model=[LinearRegression(), Lasso(), Ridge(),ElasticNet(),SVR(),DecisionTreeRegressor()]
for m in model:
    score=cross_val_score(m,x,y,cv=5,scoring='r2')
    print('score of:',m,'is:')
    print('score:',score)
    print('Mean Score:',score.mean())
    print('Standard Deviation',score.std())
    print('**********************************************************')
    print('\n')
```

```python
#Ridge Regressor model is giving best result so finding its best parameter ,using GridSearchCV
from sklearn.model_selection import GridSearchCV
parameters = {'alpha':[.0001, .01, .1, 1, 10],'fit_intercept':[True,False],'normalize':[True,False],'copy_X':[True,False],'tol':[.0001, .001, .01, 1, 1], 'random_state':list(range(0,10))}
rd=Ridge()
clf=GridSearchCV(rd,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
#Using best model with best Result

rd=Ridge(alpha=0.01,copy_X=True,fit_intercept=True,normalize=False,random_state=0,tol=0.0001)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
pred_rd=rd.predict(x_test)

rds=r2_score(y_test,pred_rd)
print('R2 score is:-',rds*100)

rdscore =cross_val_score(rd,x,y,cv=3)
rdc = rdscore.mean()
print('Cross validation Score:-',rdc*100)
```

```python
print('Error:')

print('Mean Absolute Error:',mean_absolute_error(y_test,pred_rd))
print('Mean Squared Error:',mean_squared_error(y_test,pred_rd))
print('Root Mean Square Error:',np.sqrt(mean_squared_error(y_test,pred_rd)))
```

```python
#cross Validate Score
score=cross_val_score(rd,x,y,cv=5,scoring='r2')
print("score:",score)
print('Mean score:',score.mean())
print("Standard Deviation:",score.std())
```

# Emsemble Technique

```python
from sklearn.ensemble import RandomForestRegressor 
rf=RandomForestRegressor(n_estimators=20, random_state=59)
rf.fit(x_train,y_train)
print('Score:',rf.score(x_train,y_train))
predrf=rf.predict(x_test)
print('\n')
print('Mean Absolute Error:',mean_absolute_error(y_test,predrf))
print('Mean Squared Error:',mean_squared_error(y_test,predrf))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predrf)))
print('\n')
print('r2_score:',r2_score(y_test,predrf))
```

```python
score=cross_val_score(rf,x,y,cv=5,scoring='r2')
print("score:",score)
print('Mean score:',score.mean())
print("Standard Deviation:",score.std())
```

```python
from sklearn.ensemble import AdaBoostRegressor 
adb=AdaBoostRegressor( n_estimators=100, random_state=0)
adb.fit(x_train,y_train)
print('Score:',adb.score(x_train,y_train))
predadb=adb.predict(x_test)
print('\n')
print('Mean Absolute Error:',mean_absolute_error(y_test,predadb))
print('Mean Squared Error:',mean_squared_error(y_test,predadb))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predadb)))
print('\n')
print('r2_score:',r2_score(y_test,predadb))
```

```python
score=cross_val_score(adb,x,y,cv=5,scoring='r2')
print("score:",score)
print('Mean score:',score.mean())
print("Standard Deviation:",score.std())
```
